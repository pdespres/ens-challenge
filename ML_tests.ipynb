```python
startTime = time.time()
Y = pd.read_csv('E:\Downloads\y.csv', sep=';', usecols=[1])
forest = RandomForestClassifier(n_estimators=100, oob_score=True, n_jobs=-1, random_state=42)
forest.fit(df, Y.values.ravel())
print((time.time()-startTime)/60)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import roc_auc_score

predicted = forest.predict(df)
accuracy = accuracy_score(Y, predicted)
print(f'Out-of-bag score estimate: {forest.oob_score_:.3}')
print(f'Mean accuracy score: {accuracy:.3}')
#print(roc_auc_score(Y,forest.oob_decision_function_))
```
