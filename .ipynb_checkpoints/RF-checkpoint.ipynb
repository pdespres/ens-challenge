```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
import sklearn.metrics

def transformBase():
    coltodrop = []
    df = pd.read_csv('Ctrain.csv')
    df = df.drop('Ids', axis=1)
    for i in range(0,36248,2):
        df[df.columns[i]] += df[df.columns[i + 1]] + 2
        df.rename(index=str, columns={df.columns[i]:df.columns[i][:-2]}, inplace=True)
        coltodrop.append(df.columns[i + 1])
    df.drop(coltodrop, axis=1, inplace=True)
    df.to_csv('/sgoinfre/goinfre/Perso/pdespres/ensdata/train_b0.csv', index=False)
    
%time transformBase()
df.head()
```

```python
def loadData():
    df = pd.read_csv('/sgoinfre/goinfre/Perso/pdespres/ensdata/train_b0.csv')
    y = pd.read_csv('/sgoinfre/goinfre/Perso/pdespres/ensdata/y.csv', sep=';', usecols=[1])
    return df, y
```

```python
def buildforest(train, y):
    forest = RandomForestClassifier(n_estimators=10, oob_score=True, n_jobs=-1, min_samples_leaf=50, random_state=42)
    forest.fit(train, y.values.ravel())
    print(f'Out-of-bag score estimate: {forest.oob_score_:.3}')

df, y = loadData()
print(df.shape)
%time buildforest(df, y)
```
